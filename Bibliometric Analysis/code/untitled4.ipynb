```python
import pandas as pd
df_paper1 = pd.read_csv('/content/scopus 112_bidask.csv')
df_paper2 = pd.read_csv('/content/scopus 21bid_ask.csv')
```

```python
authors_paper1 = df_paper1['Authors']
titles_paper1 = df_paper1['Title']
years_paper1 = df_paper1['Year']
keywords_paper1 = df_paper1['Author Keywords']
```

```python
authors_paper2 = df_paper2['Authors']
titles_paper2 = df_paper2['Title']
years_paper2 = df_paper2['Year']
keywords_paper2 = df_paper2['Author Keywords']
```

```python
year_stats_paper1 = years_paper1.describe()
year_stats_paper2 = years_paper2.describe()

citation_stats_paper1 = df_paper1['Cited by'].describe()
citation_stats_paper2 = df_paper2['Cited by'].describe()

print("Paper 1 - Year Statistics:")
print(year_stats_paper1)
print("\nPaper 2 - Year Statistics:")
print(year_stats_paper2)
print("\nPaper 1 - Citation Statistics:")
print(citation_stats_paper1)
print("\nPaper 2 - Citation Statistics:")
print(citation_stats_paper2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(years_paper1.value_counts().index, years_paper1.value_counts().values, label='Paper 1')
plt.bar(years_paper2.value_counts().index, years_paper2.value_counts().values, label='Paper 2')
plt.xlabel('Publication Year')
plt.ylabel('Count')
plt.title('Publication Years Distribution in Paper 1 and Paper 2')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(years_paper1, df_paper1['Cited by'], label='Paper 1')
plt.scatter(years_paper2, df_paper2['Cited by'], label='Paper 2')
plt.xlabel('Publication Year')
plt.ylabel('Citation Count')
plt.title('Publication Year vs. Citation Count in Paper 1 and Paper 2')
plt.legend()
plt.show()
```
