```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('/content/scopus 112_bidask.csv')
df2 = pd.read_csv('/content/scopus 21bid_ask.csv')

df = pd.concat([df1, df2])

# Calculate total citations for each author
author_citations = df.groupby('Authors')['Cited by'].sum()

# Sort authors by total citations in descending order
top_authors = author_citations.sort_values(ascending=False)

print('Top Authors by Total Citations:')
print(top_authors)

# Plot top authors and their total citations in a bar chart
plt.figure(figsize=(10, 6))
top_authors[:10].plot(kind='bar', color='skyblue')
plt.xlabel('Authors')
plt.ylabel('Total Citations')
plt.title('Top Authors by Total Citations')



```
